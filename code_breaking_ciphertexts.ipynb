```python
from random import *

def Metropolis(message, T, matrix, key, n = 10000):
    m = Score(message.translate(str.maketrans(CHARLIST,key)), matrix)
    for i in range(n):
        #getting random indices
        index_one = randint(0,26)
        index_two = randint(0,26)
        while index_one == index_two:
            index_one = randint(0,26)
        if index_one>index_two:
            temp = index_one
            index_one = index_two
            index_two = temp
        #Getting an alternative key and Score
        key_dash = key[:index_one]+key[index_two]+key[index_one+1:index_two]+key[index_one]+key[index_two+1:]
        m_dash = Score(message.translate(str.maketrans(CHARLIST,key_dash)), matrix)
        if m_dash > m or random() <= np.exp((m_dash-m)/T):
            m = m_dash
            key = key_dash
    return(key)

FiddleWith(ciphertexts[3],Metropolis(ciphertexts[3], 1, pair_matrix, key3))
```
